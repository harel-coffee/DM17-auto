```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_json('Fix2.json')
df.sort_index(inplace=True)
```

# uniBachelors

```python
x=df.uniBachelors.value_counts()
pd.set_option('display.max_rows', len(x))
print(x)
pd.reset_option('display.max_rows')
```

```python
df.uniBachelors.replace(u'دانشگاه','',inplace=True)
```

```python
df.uniBachelors=df.uniBachelors.str.lower()
df['uniBachelorsOLD']=df.uniBachelors.copy()
```

```python
def renamer(table):
    for i in df.index:
        f=df.get_value(i,table).strip()
        if (u'ازاد' in f) | (u'آزاد' in f)| ('azad' in f)| (u'مرکز' in f)| (u'شمال' in f)| ('iau' in f): 
            df.set_value(i,table,'azad')
        elif (u'نور' in f) | (u'پیام' in f) |('payam' in f): 
            df.set_value(i,table,'payam')
        elif (u'تربیت مدرس' in f) | ('modares' in f)| (u'مدرس' in f): 
            df.set_value(i,table,'tarbiat-modares')
        elif (u'تبریز' in f) | ('tabriz' in f): 
            df.set_value(i,table,'tabriz')
        elif (u'شیراز' in f) | ('shiraz' in f): 
            df.set_value(i,table,'shiraz')
        elif (u'مشهد' in f) | (u'فردوسی' in f)| ('mashad' in f)| ('mashhad' in f) |('ferdowsi' in f):   
            df.set_value(i,table,'mashhad')
        elif (u'شریف' in f) | ('sharif' in f):
            df.set_value(i,table,'sharif')
        elif (u'بهشتی' in f) | ('beheshti' in f)| ('sbu' in f):
            df.set_value(i,table,'beheshti')
        elif (u'امیرکبیر' in f) |(u'اميرکبير' in f) |  (u'امير کبير' in f) | (u'کبیر' in f)| ('aut' in f)| ('amir' in f)| ('kabir' in f):
            df.set_value(i,table,'amir-kabir')
        elif (u'علم و صنعت' in f) | ('iust' in f) | ('iran  science and technology' in f):
            df.set_value(i,table,'elmo-sanat')
        elif (u'علوم و تحقیقات' in f) |(u'علوم تحقیقات' in f) | ('srbiau' in f)| ('tahghighat' in f):
            df.set_value(i,table,'olom-tahghighat')
        elif (u'طوسی' in f) | (u'خواجه' in f)| ('kntu' in f)| ('khaje' in f)| ('nasir' in f)| ('tosi' in f)| ('tusi' in f)| ('toosi' in f):
            df.set_value(i,table,'toosi')
        elif (u'اصفهان' in f) | ('iut' in f)| ('isfahan' in f)| ('esfihan' in f)| ('esfahan' in f):
            df.set_value(i,table,'isfahan')
        elif (u'شاهرود' in f) | ('shahroodut' in f)| ('shahrod' in f):
            df.set_value(i,table,'shahroodut')
        elif (u'همدان' in f) | ('hamedan' in f)| ('hut' in f):
            df.set_value(i,table,'hamedan')
        elif (u'نفت' in f) | ('naft' in f)| ('put' in f)| ('oil' in f):
            df.set_value(i,table,'naft')
        elif (u'تهران' in f) | ('tehran' in f): 
            df.set_value(i,table,'tehran')
        elif (u'شريف' in f) :
            df.set_value(i,table,'sharif')
        elif (u'گیلان' in f) | ('guilan' in f)| ('gilan' in f): 
            df.set_value(i,table,'guilan')
        elif (u'ارومیه' in f) | ('urmia' in f)| ('oromiye' in f): 
            df.set_value(i,table,'urmia')
        elif (u'الزهرا' in f) | ('zahra' in f): 
            df.set_value(i,table,'al-zahra')
        elif (u'دولتی' in f) | ('dolati' in f): 
            df.set_value(i,table,'dolati')
        elif (u'سمنان' in f) | ('semnan' in f): 
            df.set_value(i,table,'semnan')
        elif (u'مازندران' in f) | ('mazandaran' in f)| ('shomal' in f):
            df.set_value(i,table,'mazandaran')
        elif (u'علامه' in f) | (u'طباطبایی' in f) | ('alame' in f):
            df.set_value(i,table,'alame-tabatabae')
        elif (u'زنجان' in f)  | ('zanjan' in f):
            df.set_value(i,table,'zanjan')
```

```python
renamer('uniBachelors')
```

```python
for i in df.index:
    f=df.get_value(i,'uniBachelors').strip()
    if f in ['tehran','sharif','beheshti','amir-kabir','tarbiat-modares','isfahan','elmo-sanat','toosi','mashhad','tabriz']:
        df.set_value(i,'highLevelBachUni',True)
    else:
        df.set_value(i,'highLevelBachUni',False)
```
