```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_json('../Data.json').T
df.sort_index(inplace=True)
df
```

```python
dg=df['apDegree'].value_counts()
dg
```

```python
dg.plot(autopct='%1.1f%%',kind='pie')
plt.show()
```

### No missing data to deal with

#### Now We're going to remove records with Degree=='Bachelors'

```python
df=df[df['apDegree']!='Bachelors']
df['apDegree'].value_counts().plot(autopct='%1.1f%%',kind='pie')
plt.show()
```

```python
df=df.copy().replace('','n/a')
df
```

```python
df.to_json('../Data2.json',date_format='utf8')
```
